```python
from scipy.interpolate import griddata

def interpolate_matrix(original_matrix, new_rows, new_cols):
    # Define the new grid coordinates
    new_x = np.linspace(0, 1, new_cols)
    new_y = np.linspace(0, 1, new_rows)
    new_x, new_y = np.meshgrid(new_x, new_y)

    # Flatten the original matrix and create corresponding coordinate grids
    old_rows, old_cols = original_matrix.shape
    old_x, old_y = np.linspace(0, 1, old_cols), np.linspace(0, 1, old_rows)
    old_x, old_y = np.meshgrid(old_x, old_y)
    original_flattened = original_matrix.flatten()

    # Interpolate using griddata
    new_matrix_flattened = griddata((old_x.flatten(), old_y.flatten()), original_flattened, (new_x, new_y), method='cubic')

    # Reshape the interpolated flattened matrix back to 2D
    new_matrix = new_matrix_flattened.reshape(new_rows, new_cols)

    return new_matrix
```
